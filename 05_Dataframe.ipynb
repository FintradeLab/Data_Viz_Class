# **3.Cleaninng**

## Column Names

```python
# Year to year
d.rename(columns = {'Year':'year'}, inplace = True)

# Country Name to country
d.rename(columns = {'Country Name':'country'}, inplace = True)

# Country Code to code
d.rename(columns = {'Country Code':'code'}, inplace = True)

# Population to pop
d.rename(columns = {'Population, total [SP.POP.TOTL]':'pop'}, inplace = True)

# Population Growth to pop_g
d.rename(columns = {'Population growth (annual %) [SP.POP.GROW]':'pop_g'}, inplace = True)

# Surface area to sur
d.rename(columns = {'Surface area (sq. km) [AG.SRF.TOTL.K2]':'sur'}, inplace = True)

# Life to lexp
d.rename(columns = {'Life expectancy at birth, total (years) [SP.DYN.LE00.IN]':'lexp'},
         inplace = True)

# Net Migration to net
d.rename(columns = {'Net migration [SM.POP.NETM]':'net'}, inplace = True)

d.head()
```

# **4. Indexing**

```python
# Duplicate the dataframe
d2 = d
```

```python
# Duplicate year as string / How to create a column
d2['year_str'] = d2['year'].astype(str)
d2.info()
```

```python
# Indexing
## Create the index
idx = pd.MultiIndex.from_product([d2['country'].unique(),range(d2.year.min(),d2.year.max()+1)],
                                 names = ['country', 'year'])

## Attach the index
d2 = d2.set_index(['country', 'year'])

d2.head()
```

# **5. Delete NaN**

```python
# Display if there are NaN
d2.isna()
```

```python
# Delete NaN
d2 = d2.dropna()
d2.info()
```

```python
# By the flies
d2 = d2.replace('..',np.NaN)
d2.head()
```

## Double axis

```python
## libraries
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
## Create a figure with secondary axis

fig3 = make_subplots(specs = [[{"secondary_y": True}]])

## Add each graph
fig3.add_trace(
    go.Scatter(x = france.year_str, y = france.net, name = 'Net Mig.'),
    secondary_y = False,
)

## Add each graph
fig3.add_trace(
    go.Scatter(x = france.year_str, y = france.pop_g, name = 'Pop Growth'),
    secondary_y = True,
)

fig3.show()
```

```python
## Create a figure with secondary axis

fig4 = make_subplots(specs = [[{"secondary_y": True}]])

## Add each graph
fig4.add_trace(
    go.Bar(x = france.year_str, y = france.net, name = 'Net Mig.'),
    secondary_y = False,
)

## Add each graph
fig4.add_trace(
    go.Scatter(x = france.year_str, y = france.pop_g, name = 'Pop Growth'),
    secondary_y = True,
)

fig4.show()
```

# **10. A little more complex graph**
